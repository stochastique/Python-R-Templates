# Testing the null hypothesis against local alternatives

Let $X_1, X_2, ...$ be i.i.d. under Laplace distributioin with density
$\frac{1}{2b}\exp\big(-\frac{|x-\theta|}{b}\big)$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Let the scale parameter be equal to 1.

```python
N = 1000 #observations
C = 1.0 #parameter for an alternative
shift, scale = 0., 1.
smpl = np.random.laplace(shift, scale, N)
```

Пусть мы хотим проверить простую гипотезу $\theta_0 = 0$ против $\theta_n \in \Big(0, \frac{C}{\sqrt n}\Big]$, где $C$ достаточно большое, чтобы быть практически полезным, и $n$ размер выборки, который может быть любым большим числом.

Сгенерируем случайное число $\theta_n$ между $\Big(0, \frac{C}{\sqrt n}\Big]$.

```python
#unknown constant signal
try:
    1 / float(C)
except ZeroDivisionError:
    print("Interval for the alternative is zero!")
try:
    theta_n = np.random.random()*C / float(np.sqrt(N))
except ZeroDivisionError:
    print("Number of observations is zero!")
```

Оно нам неизвестно (генератор случайных чисел), мы его зафиксируем в переменной $t = \sqrt{n}\theta_n$, но смотреть пока не будем.

Сгенерируем выборку $n = 1000$ из распределения Лапласа с этим параметром $\theta_n$ (вызовем через переменную, пока не знаем значение).

```python
signal1 = np.ones(N)*theta_n
try:
    z = np.arange(0,2.*np.pi,2.*np.pi/len(smpl))
except ZeroDivisionError:
    print("Number of observations is zero!")
print(len(z))
signal2 = np.sin(z) + 1.0001
plt.plot(z,signal2,z,signal1)
plt.show()
```

```python
smpl1 = signal1 + smpl
#smpl2 = signal2 + smpl
plt.plot(z,smpl,z,smpl1)
plt.show()
```
